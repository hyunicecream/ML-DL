```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.keras as optimizers
```

```python
# y = 2x^2 + 3x + 5 일 때 x, y 데이터 집합을 생성한다
x = np.array(np.arange(-5, 5, 0.1))
y = 2 * x * x + 3 * x + 5
```

```python
# 그래프를 생성한다.   
w1 = tf.Variable(1.0)
w2 = tf.Variable(1.0)
b = tf.Variable(1.0)
```

```python
def loss():
  return tf.sqrt(tf.reduce_mean(tf.square(w1 * x * x + w2 *x + b - y)))
```

```python
# Adam Optimizer 기능을 사용한다.
opt = tf.optimizers.Adam(learning_rate = 0.05)
```

```python
histloss = []
for epoch in range(300):
  opt.minimize(loss, var_list = [w1, w2, b])

  histloss.append(loss())
  if epoch % 10 == 0:
    print("epoch = %d, loss = %0.4f" % (epoch, histloss[-1]) )
```

```python
print("\n추정 결과 :")
print("w1 = %.2f" % w1.numpy())
print("w2 = %.2f" % w2.numpy())
print("b = %.2f" % b.numpy())
```

```python
plt.plot(histloss, color='red', linewidth=1)
plt.title("Loss function")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# sample data (1D)
d1 = np.random.normal(loc=1.0, scale=0.5, size=100) 
d2 = np.random.normal(loc=-3.0, scale=0.5, size=100) 
d3 = np.random.normal(loc=3.0, scale=0.5, size=100) 
data = np.concatenate([d1, d2, d3], axis=0) 
np.random.shuffle(data)
```

```python
# Kernel Density Estimator (KDE)
estimator_1 = stats.gaussian_kde(data, bw_method = 0.1)
estimator_2 = stats.gaussian_kde(data, bw_method = 0.5)
```

```python
plt.figure(figsize=(10, 5))
# estimator_1
plt.subplot(121)
X = np.arange(-8, 8, 0.1)
K = estimator_1(X)
plt.plot(X, K)

# estimator_2
plt.subplot(122)
X_ = np.arange(-8, 8, 0.1)
K = estimator_2(X_)
plt.plot(X_, K)
plt.show()
```
